```python
MAX_LINES = 3   #Global constant
MAX_BET = 100
MIN_BET = 1

# Start with collecting some user input, the deposit the user is entering as well as the bet.

# Make a function named deposit
def deposit():   # responsible for collecting user input that gets the deposit from the user
  while True:   # going to continually ask the user to enter a deposit amount until they give a valid amount.
    amount = input("What would you like to deposite? $")
    if amount.isdigit():  # .isdegit() going to tell if this is a valid whole number
      amount = int(amount)  # because input always takes input as string.
      if amount > 0:
        break   #out of the while loop.
      else:
        print("Amount must be greater than 0.")
    else:
      print("Please enter a number.")
  return amount
```

```python
#Final Compiled code:

from os import RWF_NOWAIT
# Now we need to actually run the slot machine

#import some modules
import random   # because we need to generate slot machine values kind of randomly

#we need to figure out how many items we want in each reel and how long we want the lines to be
#Here we are going to imagine that we have a three by three (3 x 3) slot machine, and that you only get a line if you get three in a row.
#We have three in a row, then you win.

MAX_LINES = 3   #Global constant
MAX_BET = 100
MIN_BET = 1

ROWS = 3
COLS = 3

#SPECIFY HOW MANY SYMBOLS ARE IN EACH OF THE REELS(columns)
symbol_count = {"A": 2,
                "B": 4,
                "C": 6,
                "D": 8} #count of symbol in each reel


symbol_value = {"A": 5,
                "B": 4,
                "C": 3,
                "D": 2}

def check_winnings(columns, lines, bet, values):
  winnings = 0
  winnings_lines = []
  for line in range(lines):
    symbol = columns[0][line]
    for column in columns:
      symbol_to_check = column[line]
      if symbol != symbol_to_check:
        break
    else:
      winnings += values[symbol] * bet
      winnings_lines.append(line + 1)
  return winnings, winnings_lines


#now we need something that is essentially going to generate what theoutcome of the slot machine was using these values.
def get_slot_machine_spin(rows, cols, symbols): #we need to generate what symbols are going to be in each column based on the frequency of symbols that we have here.
  all_symbols = []
  for symbol, symbol_count in symbols.items():
    for _ in range(symbol_count):
      all_symbols.append(symbol)

  columns = []
  for _ in range(cols):
    column = []
    current_symbols = all_symbols[:] #slice of all symbols
    for _ in range(rows):
      value = random.choice(current_symbols)
      current_symbols.remove(value)
      column.append(value)

    columns.append(column)

  return columns

#to print in a nice way
def print_slot_machine(columns):
  #changing from a horizontal view to a vertical view
  #Transposing a matrix
  for row in range(len(columns[0])):
    for i, column in enumerate(columns):
      if i != len(columns) - 1:
        print(column[row], end=" | ")
      else:
        print(column[row], end = "")

    print()

# Start with collecting some user input, the deposit the user is entering as well as the bet.

# Make a function named deposit
def deposit():   # responsible for collecting user input that gets the deposit from the user
  while True:   # going to continually ask the user to enter a deposit amount until they give a valid amount.
    amount = input("What would you like to deposite? $")
    if amount.isdigit():  # .isdegit() going to tell if this is a valid whole number
      amount = int(amount)  # because input always takes input as string.
      if amount > 0:
        break   #out of the while loop.
      else:
        print("Amount must be greater than 0.")
    else:
      print("Please enter a number.")
  return amount

# to check if the amount they are betting is within their balance

# inside of main, the next thing we want to do is, we want to collect the bet from the user
# how much they want to bet and how many line they want to bet on, and then multiply their bet amount by the number of lines
# ask the number of lines first and then ask them how much they want to bet on each line
def get_number_of_lines():   # inside here, we're going to ask to pick a number between 1 and 3, because that'll be the number of lines we have
  while True:   # going to continually ask the user to enter a deposit amount until they give a valid amount.
    lines = input("Enter the number of lines to bet on (1-" + str(MAX_LINES) + ")? ")
    if lines.isdigit():  # .isdegit() going to tell if this is a valid whole number
      lines = int(lines)  # because input always takes input as string.
      if 1 <= lines <= MAX_LINES:
        break   #out of the while loop.
      else:
        print("Enter a valid number of lines.")
    else:
      print("Please enter a number.")
  return lines



def get_bet():
  while True:   # going to continually ask the user to enter a deposit amount until they give a valid amount.
    amount = input("What would you like to bet on each line? $")
    if amount.isdigit():  # .isdegit() going to tell if this is a valid whole number
      amount = int(amount)  # because input always takes input as string.
      if MIN_BET <= amount <= MAX_BET:
        break   #out of the while loop.
      else:
        print(f"Amount must be between ${MIN_BET} - ${MAX_BET}.") # f-strings used to embed values inside string
    else:
      print("Please enter a number.")
  return amount


def spin(balance):
  lines = get_number_of_lines()
  while True:
    bet = get_bet()
    total_bet = bet * lines

    if total_bet > balance:
      print(f"You do not have enough to bet that amount, your current balance is: ${balance}")
    else:
      break
  print(f"You are betting ${bet} on {lines} lines. Total bet is equat to: ${total_bet}")

  #print(balance, lines)

  slots = get_slot_machine_spin(ROWS, COLS, symbol_count)
  print_slot_machine(slots)
  winnings, winning_lines = check_winnings(slots, lines, bet, symbol_value)
  print(f"You won ${winnings}.")
  print(f"You won on lines:", *winning_lines)
  return winnings - total_bet


def main():   # if we end our program, we can call this function again and then it will rerun the program
  balance = deposit()
  while True:
    print(f"Current balance is ${balance}")
    answer = input("Press enter to play (q to quit).")
    if answer == "q":
      break
    balance += spin(balance)

  print(f"You left with ${balance}")


main()
```
